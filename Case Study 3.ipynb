```python
import pandas as pd
import numpy as np

terror=pd.read_csv("C://Users//Administrator//Documents//Python//Python Assignments//DataSets//terror.csv//terror.csv",sep=",",header=0,encoding='ISO-8859-1')

terror.head()
```

```python
# How many attacks happened in India?

terror[terror.country_txt=="India"].shape[0]
```

```python
# How many attacks happened in India and upto 3 people were killed? 

terror[(terror.country_txt=="India") & (terror.nkill<=3)].shape[0]
```

```python
# Extract the city and summary for attacks above

terror[(terror.country_txt=="India") & (terror.nkill<=3)][['country_txt','city','summary']]
```

```python
# In a single terror incident in India, find out top 5 cities by number killed

terror[terror.country_txt=="India"][['city','nkill']].sort_values(by='nkill',ascending=False).head()
```

```python
# In a single terror incident in India, find out top 5 cities by number killed and wounded 

terror[terror.country_txt=="India"][['city','nkill','nwound']].sort_values(by='nwound',ascending=False).head()
```

```python
# How many attacks were successful that were suicide attacks?

terror[(terror.success==1) & (terror.suicide==1)].shape[0]
```

```python
# label all the incidents where the number killed was more than 5 as severe. 

terror['severe']=['severe' if i>5.0 else 'not severe' for i in terror.nkill]
terror['severe']
```

```python
# write a function to label an incident that was both successful and suicidal

def func(x):
            x['Success/Failure']=["Suicide_success" if (x.iloc[i].success==1 and x.iloc[i].suicide==1) else "Suicide_failure" for i in list(range(0,len(x)))]
            return x['Success/Failure']

func(terror)
```

```python
# Create a new category representing if the incident occured in Afghanistan, Pakistan or India as one level
#  of the category and all the other countries as another level

terror['Category']=["Asia" if terror.iloc[i].country_txt=="Afghanistan" or terror.iloc[i].country_txt=="Pakistan" or terror.iloc[i].country_txt=="India" else "Other" for i in list(range(0,len(terror)))] 
terror.Category.head()
```

```python
# How many incidents happened in Af-Pak-India vs ROW?


print("No. of Incidents in Afghanisthan, Pakistan and India is: ", (terror[terror.Category=='Asia'].shape[0]))
len(terror[terror.Category=='Other'])
round((terror[terror.Category=='Asia'].shape[0])*100/(terror[terror.Category=='Other'].shape[0]),2)
```

```python
# List the number of suicides attacks and average kills by Af-Pak-India vs ROW. Rename columns in the output as Average_Kills
#  and Number_Incidents.

terror[terror.suicide == 1].groupby('Category')[['nkill']].agg(['count','mean']).rename(columns = {'count' : "Number_Incidents",'mean':"Average_kills"})
```
