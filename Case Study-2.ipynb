```python
import pandas as pd
import csv
import numpy as np

st=pd.read_csv("C://Users//Administrator//Documents//Python//Python Assignments//DataSets//Store.csv",sep=",",header=0,encoding='latin')
st.head()
```

```python
# â€¢	How many unique cities are the orders being delivered to
st.City.unique()
```

```python
# â€¢	What is the total quantity sold in the East Region?

st[st.Region=='East'].Quantity.count()
```

```python
# â€¢	Find the sum of the quantity sold in the East Region
st[st.Region=='East'].Quantity.sum()
```

```python
# â€¢	In the south region sort the sales in decreasing order
st[st.Region=='South'].sort_values(by='Sales',ascending=False)
```

```python
# â€¢	Find the mean of quantity for every region
st.groupby('Region').Quantity.mean()
```

```python
# â€¢	Find the mean of sales for every category
st.groupby('Category').Sales.mean()
```

```python
# â€¢	Find the max, min, sum of sales and profit for every category

#st.groupby('Category').agg({'Sales': ['max','min','sum'],'Profit': ['max','min','sum']})
st.groupby('Category')[['Sales','Profit']].agg(['max','min','sum'])
```

```python
#â€¢	Find sum of sales and max profit for every segment
st.groupby('Segment')[['Sales','Profit']].agg(['sum','max'])
```

```python
#â€¢	For every segment find the mean of the discount

st.groupby('Segment').Discount.mean()
```

```python
#â€¢	For every segment find the most profitable customers 

#st.sort_values(by='Profit',ascending=False)
def get_cust(pro):
    pro = pro.max()
    return st[st.Profit == pro]['Customer ID']
s = st.groupby('Segment').agg({'Profit':[get_cust,'max']})
s.Profit.sort_values('max', ascending = False)
```

```python
# â€¢	What are the top 5 categories that give maximum profit?

#hg = st.groupby('Sub-Category').max()['Profit'].sort_values(ascending = False).head()
#hg

st.groupby('Sub-Category')['Profit'].max().sort_values(ascending=False).head()
```

```python
#â€¢	What is the Total Sales, Quantity, Discount, Profit across Total US.

st[['Sales','Quantity','Discount','Profit']].agg({np.sum})
```

```python
#â€¢	How many times has it taken more than 5 days from placing an order to shipping
st['Days']=pd.to_datetime(st['Ship Date'])-pd.to_datetime(st['Order Date'])
st['D5']=(Days > '5 days')
st.D5.sum()
```

```python
#â€¢	Find the total number of orders in every category which has been shipped with a duration > 5 days
st.groupby('Category').D5.sum()
```

```python
#â€¢	Whatâ€™s the percentage of items which has been shipped within 5 days

st['Dless5']=(Days<'5 days')
round((st.Dless5.sum())/len(st)*100,3)
```

```python
#â€¢	Whatâ€™s the percentage of items which has been shipped after 5 days
round((st.D5.sum())/len(st)*100,3)
```
